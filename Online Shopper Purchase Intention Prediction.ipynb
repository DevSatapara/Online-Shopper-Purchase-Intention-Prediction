# 1) Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc

logistic_reg = LogisticRegression()

# Fit the model
logistic_reg.fit(x_train, y_train)
    
# Predict probabilities for training and validation sets
y_train_pred_prob = logistic_reg.predict_proba(x_train)[:, 1]
y_val_pred_prob = logistic_reg.predict_proba(x_val)[:, 1]

# Compute ROC curve and AUC for training set
fpr_train, tpr_train, _ = roc_curve(y_train, y_train_pred_prob)
roc_auc_train = auc(fpr_train, tpr_train)

# Compute ROC curve and AUC for validation set
fpr_val, tpr_val, _ = roc_curve(y_val, y_val_pred_prob)
roc_auc_val = auc(fpr_val, tpr_val)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr_train, tpr_train, color='blue', lw=2, label='Training ROC curve (AUC = %0.2f)' % roc_auc_train)
plt.plot(fpr_val, tpr_val, color='red', lw=2, label='Validation ROC curve (AUC = %0.2f)' % roc_auc_val)
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()
```

```python
from sklearn.metrics import accuracy_score

learning_rates = [0.0001, 0.001, 0.01, 0.1, 1, 10]

# Initialize lists to store accuracies
accuracies_train = []
accuracies_val = []

# Iterate over different learning rates
for lr in learning_rates:
    # Initialize the logistic regression model
    logistic_reg = LogisticRegression(penalty='l2',C=1/lr,solver='lbfgs')

    # Fit the model
    logistic_reg.fit(x_train, y_train)
    
    # Evaluate accuracy on training set
    train_accuracy = logistic_reg.score(x_train, y_train)
    accuracies_train.append(train_accuracy)
    
    # Evaluate accuracy on validation set
    val_accuracy = logistic_reg.score(x_val, y_val)
    accuracies_val.append(val_accuracy)

# Plot learning rate vs accuracy
plt.figure(figsize=(10, 6))
plt.plot(learning_rates, accuracies_train, label='Training Accuracy', marker='o')
plt.plot(learning_rates, accuracies_val, label='Validation Accuracy', marker='o')
plt.xlabel('Learning Rate')
plt.ylabel('Accuracy')
plt.title('Learning Rate vs Accuracy for Logistic Regression')
plt.legend()
plt.xscale('log')
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc

learning_rates = [0.0001, 0.001, 0.01, 0.1, 1, 10]

# Initialize lists to store AUCs
aucs_train = []
aucs_val = []

# Iterate over different learning rates
for lr in learning_rates:
    # Initialize the logistic regression model
    logistic_reg = LogisticRegression(penalty='l2',C=1/lr,solver='lbfgs')

    # Fit the model
    logistic_reg.fit(x_train, y_train)
    
    # Predict probabilities for training set
    y_train_pred_prob = logistic_reg.predict_proba(x_train)[:, 1]
    fpr_train, tpr_train, _ = roc_curve(y_train, y_train_pred_prob)
    roc_auc_train = auc(fpr_train, tpr_train)
    aucs_train.append(roc_auc_train)
    
    # Predict probabilities for validation set
    y_val_pred_prob = logistic_reg.predict_proba(x_val)[:, 1]
    fpr_val, tpr_val, _ = roc_curve(y_val, y_val_pred_prob)
    roc_auc_val = auc(fpr_val, tpr_val)
    aucs_val.append(roc_auc_val)

# Plot learning rate vs AUC
plt.figure(figsize=(10, 6))
plt.plot(learning_rates, aucs_train, label='Training AUC', marker='o')
plt.plot(learning_rates, aucs_val, label='Validation AUC', marker='o')
plt.xlabel('Learning Rate')
plt.ylabel('AUC')
plt.title('Learning Rate vs AUC for Logistic Regression')
plt.legend()
plt.xscale('log')
plt.show()
```

# 2) Decision Tree

```python
def entropy(labels):
    _, label_occurence = np.unique(labels, return_counts=True) # Calculating the total occurence of the label in the data
    probability = label_occurence / len(labels) # calulating the probability of the label
    entropy_value = -np.sum(probability * np.log2(probability)) # calculating the entropy of the label
    return entropy_value

def information_gain(feature, splits):
    feature_entropy = entropy(feature) # calculating the entropy of the feature
    total_samples = sum(len(split) for split in splits) # finding the total splits of the features
    split_entropy = sum((len(split) / total_samples) * entropy(split) for split in splits) # calulating the entropy of the split
    information_gain_value = feature_entropy - split_entropy #calculating the information gain
    return information_gain_value

# Calculate Information Gain for each feature in the training set
information_gains = {}
for feature in x_train.columns:
    unique_values = x_train[feature].unique() # Calculating the unique values in the feature
    splits = [y_train[x_train[feature] == value] for value in unique_values] # Finding splits of the feature
    information_gains[feature] = information_gain(y_train, splits) # Calucalting the information gain of the feature

print("Information Gains:")
for feature, gain in information_gains.items():
    print(f"{feature}: {gain}")

max_gain_feature = max(information_gains, key=information_gains.get)
max_gain = information_gains[max_gain_feature]

print(f"\n\n The feature with maximum gain is '{max_gain_feature}' with a gain of {max_gain}")
```

```python
decision_tree_classifier = DecisionTreeClassifier(random_state=42,max_leaf_nodes=20)
decision_tree_classifier.fit(x_train,y_train)

plt.figure(figsize=(30, 20))
plot_tree(decision_tree_classifier, feature_names=X.columns, class_names=['Revenue Not Generated','Revenue Generated'],filled=True, rounded=True, fontsize=10)
plt.title("Decision Tree")
plt.show()
```

```python
# Define your performance metrics including accuracy, precision, recall and F1-scorre
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def calculate_metrics(true_labels, predicted_labels):
    accuracy = accuracy_score(true_labels, predicted_labels)
    precision = precision_score(true_labels, predicted_labels)
    recall = recall_score(true_labels, predicted_labels)
    f1 = f1_score(true_labels, predicted_labels)
    return accuracy,precision,recall,f1

predict = decision_tree_classifier.predict(x_test)
acc,pre,rec,f1_s = calculate_metrics(y_test,predict)
performance_metrics = {
    'Accuracy': {acc},
    'Precision': {pre},
    'Recall': {rec},
    'F1 Score': {f1_s}
}
for metric, value in performance_metrics.items():
    print(f"{metric}: {value}") # Printing the model performance parameters
```

```python
from sklearn.model_selection import GridSearchCV

# Define a range of max_leaf_nodes values to search
param_grid = {'max_leaf_nodes': list(range(5, 25)),
              'criterion' : ['entropy','gini','log_loss'],
              'max_depth': list(range(1,10))}

# Run grid search with cross-validation
grid_search = GridSearchCV(DecisionTreeClassifier(random_state=42), param_grid, cv=5)
grid_search.fit(x_train, y_train) # Note testing set cannot be used

# Get the best estimator with the optimal max_leaf_nodes parameter
print(grid_search.best_estimator_)

# Plot the pruned decision tree
best_estimator = grid_search.best_estimator_

# Plot the pruned decision tree
plt.figure(figsize=(30, 20))
plot_tree(best_estimator, feature_names=new_df.columns,  class_names=['Revenue Not Generated','Revenue Generated'], filled=True, rounded=True, fontsize=10)
plt.title("Pruned Decision Tree for Dataset")
plt.show()
```

```python
y_pred_best = best_estimator.predict(x_test)

accuracy_best = accuracy_score(y_test, y_pred_best)
print("Accuracy of the best estimator:", accuracy_best)
```

**Insights:- Decision Tree Model performs well with updating the hyperparameters with the help of GridSearch CV and getting the accuracy of 89%.**

```python
# Define the parameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': list(range(1,10)),
    'min_samples_split': list(range(1,10)),
    'min_samples_leaf': list(range(1,5)),
    'bootstrap': [True, False]
}

# Initialize the RandomForestClassifier
rf = RandomForestClassifier()

# Perform GridSearchCV
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=3, scoring='accuracy', n_jobs=-1)
grid_search.fit(x_train, y_train)

# Print the best parameters found
print("Best parameters:", grid_search.best_params_)

# Evaluate the model on the test set using the best parameters
best_model = grid_search.best_estimator_
test_accuracy = best_model.score(x_test, y_test)
print("Accuracy on test set:", test_accuracy)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier

accuracies = []
k_values = range(1, 21)  # Trying K values from 1 to 20

# Calculate accuracy for each value of K
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
    y_pred = knn.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)

best_k = k_values[np.argmax(accuracies)]
best_accuracy = max(accuracies)

# Plot the elbow curve
plt.plot(k_values, accuracies, marker='o')
plt.title(f'Elbow Curve for KNN')
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Accuracy')
plt.xticks(k_values)
plt.show()
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors=best_k)

# Train the KNN classifier on the training data
knn_classifier.fit(x_train, y_train)

# Make predictions on the test data
y_pred_knn = knn_classifier.predict(x_test)

# Calculate and print the accuracy of the KNN classifier
accuracy_knn = accuracy_score(y_test, y_pred_knn)
print("Accuracy of the KNN classifier:", accuracy_knn)
```
